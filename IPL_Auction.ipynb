```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.iplt20.com/auction/2020"
r= requests.get(url)
soup = BeautifulSoup(r.text,"lxml")
# print(soup)
table= soup.find_all("table", class_ = "ih-td-tab auction-tbl")
# print(table)
head = soup.find_all("th",class_="skip-filter")
print(head)
```

```python
header= soup.find("tbody",id = "pointsdata")
tables = []
for i in header:
    tab = i.text.split("\n")
#     if tab[1]!="":
    tables.append(tab)
# print(tables)
df = pd.DataFrame(tables)
df
```
