```python
l = []
while True:
    try:
        c = int(input('''
        1 push or enqueue elements
        2 pop or dequeue elements
        3 show front or first elements
        4 show last or rear elements
        5 Display stack
        6 exit
        '''))
        if c==1:
            try:
                n = int(input("ENter nth time you want to insert data:-"))
                for i in range(n):
                    m = input("Enter the value:-")
                    l.append(m)
                    print(l)
            except:
                print("Enter integer number")
        elif c==2:
            if len(l)==0:
                print("empty queue")
            else:
                del l[0]
                print(l)
        elif c ==3:
            if len(l)==0:
                print("Empty queue")
            else:
                print("the first or front value is:",l[0])
        elif c==4:
            if len(l)==0:
                print("Empty queue")
            else:
                print("the last or rear queue's value is:-",l[-1])
        elif c==5:
            print("Display queue:-",l)
        elif c==6:
            break
    except:
            print("enter valid number")
```
