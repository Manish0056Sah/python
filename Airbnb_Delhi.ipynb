```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
Names =[]
Prices=[]
Reviews =[]

np = soup.find("a",class_="l1ovpqvx c1ytbx3a dir dir-ltr").get("href")  #np = next page
cnp = "https://www.airbnb.com/"+np
for i in range(1,6):
    url ="https://www.airbnb.com/s/Delhi--India/homes?adults=1&place_id=ChIJL_P_CXMEDTkRw0ZdG-0GVvw&refinement_paths%5B%5D=%2Fhomes"
    r =requests.get(url)
    # print(r)

    soup = BeautifulSoup(r.text,"lxml")     #using soup the whole pages are inserted
    box = soup.find("div",class_="f1ym1fv dir dir-ltr")
    names = box.find_all("div",class_="t1jojoys dir dir-ltr")
    # print(Names)
    for i in names:
        d =i.text
        Names.append(d)
    # print(len(Names))
#     subtitle = box.find_all("div",class_="fb4nyux s1cjsi4j dir dir-ltr")
#     for i in subtitle:
#         s = i.text
#         Subtitle.append(s)
#     print(len(Subtitle))
#     prices =box.find_all("div",class_="pquyp1l dir dir-ltr")
    # print(prices)
    for i in prices:
        p = i.text
        Prices.append(p)
    # print(Prices)
    # print(Subtitle)

    # bed_no = box.find_all("div",class_="fb4nyux s1cjsi4j dir dir-ltr")
    # print(len(bed_no))
    review = box.find_all("span",class_="t1a9j9y7 r4a59j5 dir dir-ltr")
    # print(review)
    for i in review:
        r = i.text
        Reviews.append(r)
# print(Reviews)
df = pd.DataFrame({"Hotel_Name":Names,"Hotel_Prices":Prices,"Hotel_Review":Reviews})
print(df)

df.to_csv("Airbnb_Delhi.csv")

# print(cnp)
# url = cnp          #  This process will work when all next page have diffrent urls here only page_no. is changed
# soup = requests.get(url) 
# print(soup)
# soup = BeautifulSoup(r.text,"lxml")     #using soup the whole pages are inserted
```
