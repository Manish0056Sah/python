```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
Results = []
View = [] 
Notes = []
Type = []
for i in range(1,3):
    url = "https://catalog.data.gov/dataset/?page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text,"lxml")

    results = soup.find_all('h3', class_="dataset-heading")

    # Iterate over the individual elements in the ResultSet
    for result in results:
        # Now you can call the find_all() method on each element
        sub_results = result.find_all('a')
        # Process the sub_results as needed
        for sub_result in sub_results:
            Results.append(sub_result.text)
    print(len(Results))
    for view in results:
        s_view = view.find_all("span")
        for s_vie in s_view:
            View.append(s_vie.text.split())
    print(len(View))


    notes = soup.find_all("p",class_="dataset-organization")
    # print(notes)

    for n in notes:
        nn=n.text.split()
        Notes.append(nn)
    print(len(Notes))

    type = soup.find_all("span",class_="organization-type")
    for i in type:
        m = i.find_all("span")
        for n in m:
            Type.append(n.text)
    print(len(Type))
    # D_format =[]
    # d_format =soup.find_all("div",class_="dataset-content")
    # for i in d_format:
    #     m = i.find_all("ul")
    #     for n in m:    
    # #         o =n.find_all("a")
    # #         for p in o:
    #             D_format.append(n.text.split())
    # print(len(D_format))
df = pd.DataFrame({"Header":Results,"Views":View,"Notes":Notes,"Org_Type":Type})
df.to_csv("Gov_data.csv")
df
```
