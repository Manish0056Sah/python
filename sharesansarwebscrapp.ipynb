```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.sharesansar.com/live-trading', 'html.parser')
soup = requests.get(url)
soup
```

```python
raw_data = soup.text
raw_data =BeautifulSoup(raw_data, "lxml")	#this data is in string format so we have to change this 
raw_data
```

```python
table_code = raw_data.table
table_code
```

```python
tags=table_code.find_all('tr')
tags
```

```python
for j in tags:
    l = []
    td_tags = j.find_all('td')
    ll = list(map(lambda x:x.text.split(), td_tags))
    print(ll)
for tag in ll:
    y=tag.text.split('\n') 
#     if y[1]!="":
    ll.append(y)
ll
```

```python
data = []
for tag in tags[1:]:
    y=tag.text.split('\n') 
#     if y[1]!="":
    data.append(y[2:])
data
```

```python
import csv
file = open('sharesansar_data.csv','w')
x= csv.writer(file)
x.writerows(data)
file.close()
```
