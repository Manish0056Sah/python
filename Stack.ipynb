```python
l = []

while True:
    try:
        c = eval(input('''
        1 push elements
        2 pop elements
        3 peek element
        4 Display Stack
        5 exit

        '''))
        if c==1:
            try:
                n = int(input("Enter nth time you want to  insert data:-"))
                for i in range(n):
                    m = input("Enter the data :-")
                    l.append(m)
                    print(l)
            except:
                print("enter interger value") 
        elif c==2:
            if len(l)==0:
                print("Empty stack")
            else:
                p = l.pop()
                print(p)
                print(l)
        elif c==3:
            if len(l)==0:
                print("Empty stack")
            else:
                print("last stack value is:",l[-1])
        elif c==4:
            print("Stack is:",l)
        elif c==5:
            break
        else:
          print("Enter valid number") 
    except:
        print("Enter valid number")
```
